```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, LSTM, Dense, Dropout, BatchNormalization, Flatten, Concatenate, Reshape
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Load data
with h5py.File('/content/Bspectrum_sensing_dataset.mat', 'r') as f:
    energy = np.array(f['energy_matrix']).transpose(2, 1, 0)           # [SNR, trials, SUs]
    cov = np.array(f['covariance_matrices']).transpose(1, 0, 2, 3)     # [SNR, trials, 5, 5]
    labels = np.array(f['target_vector']).astype(int).T                # [SNR, trials]
    snrs = np.array(f['SNR_dB']).flatten()

# Data dimensions
num_SNR, num_trials, num_SUs = energy.shape
_, _, cov_dim1, cov_dim2 = cov.shape

print(f"Dataset dimensions:")
print(f"Energy matrix: {energy.shape}")
print(f"Covariance matrices: {cov.shape}")
print(f"Labels: {labels.shape}")
print(f"SNR values: {snrs}")

def create_sequences(energy_data, cov_data, labels, seq_len, stride=1):
    num_SNR, num_trials, num_SUs = energy_data.shape
    num_sequences_per_snr = (num_trials - seq_len) // stride + 1

    # Initialize arrays
    X_energy = np.zeros((num_SNR * num_sequences_per_snr, seq_len, num_SUs))
    # Initialize X_cov with a complex data type, but allow for structured input
    X_cov = np.zeros((num_SNR * num_sequences_per_snr, seq_len, cov_dim1 * cov_dim2), dtype=complex)
    Y = np.zeros(num_SNR * num_sequences_per_snr)
    SNR_levels = np.zeros(num_SNR * num_sequences_per_snr)

    idx = 0
    for snr_idx in range(num_SNR):
        for t in range(0, num_trials - seq_len + 1, stride):
            # Extract sequences
            X_energy[idx] = energy_data[snr_idx, t:t+seq_len, :]

            # Flatten and store covariance matrices, handling structured data
            for seq_pos in range(seq_len):
                # Convert structured array to complex using 'real' and 'imag' fields
                cov_complex = cov_data[snr_idx, t+seq_pos]['real'] + 1j * cov_data[snr_idx, t+seq_pos]['imag']
                X_cov[idx, seq_pos] = cov_complex.flatten()

            # Use the label of the last observation in the sequence
            Y[idx] = labels[snr_idx, t+seq_len-1]

            # Store SNR level for this sequence (for analysis)
            SNR_levels[idx] = snrs[snr_idx]

            idx += 1

    return X_energy, X_cov, Y, SNR_levels

# Create sequences for model input
sequence_length = 5  # Use 5 consecutive trials as one sequence
X_energy_seq, X_cov_seq, Y_seq, sequence_SNRs = create_sequences(
    energy, cov, labels, sequence_length, stride=1)

print(f"After sequence creation:")
print(f"X_energy_seq shape: {X_energy_seq.shape}")
print(f"X_cov_seq shape: {X_cov_seq.shape}")
print(f"Y_seq shape: {Y_seq.shape}")

# Split data into training, validation, and test sets (70%, 15%, 15%)
# Ensure balanced representation across SNR levels
indices = np.arange(len(Y_seq))
X_train_indices, X_test_indices, y_train_indices, y_test_indices = train_test_split(
    indices, indices, test_size=0.3, stratify=sequence_SNRs, random_state=42)

X_val_indices, X_test_indices, y_val_indices, y_test_indices = train_test_split(
    X_test_indices, y_test_indices, test_size=0.5, stratify=sequence_SNRs[X_test_indices], random_state=42)

# Training set
X_energy_train = X_energy_seq[X_train_indices]
X_cov_train = X_cov_seq[X_train_indices]
Y_train = Y_seq[X_train_indices]
SNR_train = sequence_SNRs[X_train_indices]

# Validation set
X_energy_val = X_energy_seq[X_val_indices]
X_cov_val = X_cov_seq[X_val_indices]
Y_val = Y_seq[X_val_indices]
SNR_val = sequence_SNRs[X_val_indices]

# Test set
X_energy_test = X_energy_seq[X_test_indices]
X_cov_test = X_cov_seq[X_test_indices]
Y_test = Y_seq[X_test_indices]
SNR_test = sequence_SNRs[X_test_indices]

def build_cnn_model(sequence_length, num_features, cov_features):
    # Energy input branch
    energy_input = Input(shape=(sequence_length, num_features), name='energy_input')
    x1 = Conv1D(filters=64, kernel_size=3, padding='same', activation='relu')(energy_input)
    x1 = BatchNormalization()(x1)
    x1 = Conv1D(filters=128, kernel_size=3, padding='same', activation='relu')(x1)
    x1 = BatchNormalization()(x1)
    x1 = Flatten()(x1)  # Flatten the output for the Dense layer

    # Covariance input branch
    cov_input = Input(shape=(sequence_length, cov_features), name='cov_input')
    x2 = Conv1D(filters=64, kernel_size=3, padding='same', activation='relu')(cov_input)
    x2 = BatchNormalization()(x2)
    x2 = Conv1D(filters=128, kernel_size=3, padding='same', activation='relu')(x2)
    x2 = BatchNormalization()(x2)
    x2 = Flatten()(x2)  # Flatten the output for the Dense layer

    # Merge branches
    merged = Concatenate()([x1, x2])

    # Dense layers for classification
    x = Dense(128, activation='relu')(merged)
    x = BatchNormalization()(x)
    x = Dropout(0.4)(x)
    x = Dense(64, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)
    output = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=[energy_input, cov_input], outputs=output)
    return model

# Create and compile model
model = build_cnn_model(sequence_length, num_SUs, cov_dim1 * cov_dim2)
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC()]
)

# Display model summary
model.summary()

# Train model with early stopping and learning rate reduction
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss', patience=10, restore_best_weights=True)
lr_scheduler = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6)

# Train the model
history = model.fit(
    [X_energy_train, X_cov_train], Y_train,
    validation_data=([X_energy_val, X_cov_val], Y_val),
    epochs=50,
    batch_size=64,
    callbacks=[early_stopping, lr_scheduler],
    verbose=1
)

# Evaluate model
test_loss, test_acc, test_auc = model.evaluate([X_energy_test, X_cov_test], Y_test)
print(f"Test accuracy: {test_acc:.4f}")
print(f"Test AUC: {test_auc:.4f}")

# Plot training history - accuracy and loss curves
plt.figure(figsize=(16, 6))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.savefig('learning_curves.png')
plt.show()

# Make predictions on test set
y_pred_prob = model.predict([X_energy_test, X_cov_test])
y_pred = (y_pred_prob > 0.5).astype(int).flatten()

# Calculate confusion matrix
cm = confusion_matrix(Y_test, y_pred)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.savefig('confusion_matrix.png')
plt.show()

# Calculate probability of detection vs SNR
def calculate_pd_vs_snr(y_true, y_pred, snr_values, all_snrs):
    unique_snrs = np.sort(np.unique(all_snrs))
    pd_values = []

    for snr in unique_snrs:
        # Find indices for this SNR
        indices = np.where(snr_values == snr)[0]

        # Filter for only positive cases (signal present)
        positive_indices = np.where((y_true[indices] == 1))[0]

        if len(positive_indices) > 0:
            # Calculate probability of detection (true positive rate)
            true_positives = np.sum((y_true[indices] == 1) & (y_pred[indices] == 1))
            pd = true_positives / len(positive_indices)
            pd_values.append(pd)
        else:
            pd_values.append(None)  # No positive cases for this SNR

    return unique_snrs, pd_values

# Calculate Pd vs SNR
unique_snrs, pd_values = calculate_pd_vs_snr(Y_test, y_pred, SNR_test, snrs)

# Plot Pd vs SNR
plt.figure(figsize=(10, 6))
plt.plot(unique_snrs, pd_values, 'o-', markersize=8)
plt.grid(True)
plt.xlabel('SNR (dB)')
plt.ylabel('Probability of Detection (Pd)')
plt.title('Probability of Detection vs SNR')
plt.savefig('pd_vs_snr.png')
plt.show()

# Calculate false alarm rate
def calculate_pfa_vs_snr(y_true, y_pred, snr_values, all_snrs):
    unique_snrs = np.sort(np.unique(all_snrs))
    pfa_values = []

    for snr in unique_snrs:
        # Find indices for this SNR
        indices = np.where(snr_values == snr)[0]

        # Filter for only negative cases (signal absent)
        negative_indices = np.where((y_true[indices] == 0))[0]

        if len(negative_indices) > 0:
            # Calculate probability of false alarm
            false_positives = np.sum((y_true[indices] == 0) & (y_pred[indices] == 1))
            pfa = false_positives / len(negative_indices)
            pfa_values.append(pfa)
        else:
            pfa_values.append(None)  # No negative cases for this SNR

    return unique_snrs, pfa_values

# Calculate Pfa vs SNR
unique_snrs, pfa_values = calculate_pfa_vs_snr(Y_test, y_pred, SNR_test, snrs)

# Plot Pfa vs SNR
plt.figure(figsize=(10, 6))
plt.plot(unique_snrs, pfa_values, 'o-', markersize=8, color='red')
plt.grid(True)
plt.xlabel('SNR (dB)')
plt.ylabel('Probability of False Alarm (Pfa)')
plt.title('Probability of False Alarm vs SNR')
plt.savefig('pfa_vs_snr.png')
plt.show()

# Plot ROC curves for different SNR levels
from sklearn.metrics import roc_curve, auc

plt.figure(figsize=(12, 8))

# Get a subset of SNR values to plot (to avoid overcrowding)
snr_plot_indices = np.linspace(0, len(unique_snrs)-1, min(6, len(unique_snrs))).astype(int)
selected_snrs = [unique_snrs[i] for i in snr_plot_indices]

for snr in selected_snrs:
    # Find indices for this SNR
    indices = np.where(SNR_test == snr)[0]

    if len(indices) > 0:
        # Calculate ROC curve
        fpr, tpr, _ = roc_curve(Y_test[indices], y_pred_prob[indices])
        roc_auc = auc(fpr, tpr)

        # Plot ROC curve
        plt.plot(fpr, tpr, lw=2, label=f'SNR = {snr} dB (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Different SNR Levels')
plt.legend(loc="lower right")
plt.grid(True)
plt.savefig('roc_curves.png')
plt.show()

# Create a summary of results
print("\n--- Model Performance Summary ---")
print(f"Test Accuracy: {test_acc:.4f}")
print(f"Test AUC: {test_auc:.4f}")
print("\nConfusion Matrix:")
print(cm)

```
